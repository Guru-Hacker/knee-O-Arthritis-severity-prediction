```python
# Cell 1: Fixed Imports
import numpy as np
import pandas as pd
import os
import time
import matplotlib.pyplot as plt
import cv2
import seaborn as sns
import shutil
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import (Dense, Activation, Dropout, Conv2D, MaxPooling2D, 
                                      BatchNormalization, GlobalAveragePooling2D, Input,
                                      Flatten, concatenate, GlobalMaxPooling2D)
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
from tensorflow.keras import regularizers
from tensorflow.keras.models import Model
from tensorflow.keras import backend as K
from tensorflow.keras.applications import VGG16
from tensorflow.keras.preprocessing import image
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint, LearningRateScheduler

# Define save directory for Kaggle working outputs
save_dir = '/kaggle/working'
os.makedirs(save_dir, exist_ok=True)

# Cell 2: Set up Kaggle paths
train_path = '/kaggle/input/knee-osteoarthritis-dataset-with-severity/train'
test_path = '/kaggle/input/knee-osteoarthritis-dataset-with-severity/test'
valid_path = '/kaggle/input/knee-osteoarthritis-dataset-with-severity/val'
list_of_classes=['Healthy', 'Doubtful', 'Minimal', 'Moderate', 'Severe']

# Cell 3: Data analysis
for d in [train_path, test_path, valid_path]:
    filepaths = []
    labels=[]
    classlist=os.listdir(d)
    for klass in classlist:
        intklass=int(klass)
        label=list_of_classes[intklass]
        classpath=os.path.join(d, klass)
        flist=os.listdir(classpath)
        for f in flist:
            fpath=os.path.join(classpath,f)
            filepaths.append(fpath)
            labels.append(label)
    Fseries=pd.Series(filepaths, name='filepaths')
    Lseries=pd.Series(labels, name='labels')
    pdf=pd.concat([Fseries, Lseries], axis=1)
    if d == test_path:
        test_df=pdf
    elif d == valid_path:
        valid_df=pdf
    else:
        train_df=pdf
        
print('train_df length: ', len(train_df), '  test_df length: ', len(test_df), '  valid_df length: ', len(valid_df))

classes=sorted(list(train_df['labels'].unique()))
class_count = len(classes)
print('The number of classes in the dataset is: ', class_count)
groups=train_df.groupby('labels')
print('{0:^30s} {1:^13s}'.format('CLASS', 'IMAGE COUNT'))
countlist=[]
classlist=[]
for label in sorted(list(train_df['labels'].unique())):
    group=groups.get_group(label)
    countlist.append(len(group))
    classlist.append(label)
    print('{0:^30s} {1:^13s}'.format(label, str(len(group))))

# Cell 4: Data trimming function
def trim(df, max_samples, min_samples, column):
    df=df.copy()
    groups=df.groupby(column)
    trimmed_df = pd.DataFrame(columns = df.columns)
    for label in df[column].unique():
        group=groups.get_group(label)
        count=len(group)
        if count > max_samples:
            sampled_group=group.sample(n=max_samples, random_state=123,axis=0)
            trimmed_df=pd.concat([trimmed_df, sampled_group], axis=0)
        else:
            if count>=min_samples:
                sampled_group=group
                trimmed_df=pd.concat([trimmed_df, sampled_group], axis=0)
    print('after trimming, the maximum samples in any class is now ',max_samples, ' and the minimum samples in any class is ', min_samples)
    return trimmed_df

max_samples=600  # Increased for better training
min_samples=173
column='labels'
train_df= trim(train_df, max_samples, min_samples, column)

# Cell 5: Enhanced Data balancing function
def balance(df, n, working_dir, img_size):
    df=df.copy()
    print('Initial length of dataframe is ', len(df))
    aug_dir=os.path.join(working_dir, 'aug')
    if os.path.isdir(aug_dir):
        shutil.rmtree(aug_dir)
    os.makedirs(aug_dir, exist_ok=True)
    for label in df['labels'].unique():
        dir_path=os.path.join(aug_dir,label)
        os.makedirs(dir_path, exist_ok=True)
    total=0
    # Enhanced augmentation for better diversity
    gen=ImageDataGenerator(
        horizontal_flip=True, 
        rotation_range=25, 
        width_shift_range=.25,
        height_shift_range=.25, 
        zoom_range=.25,
        brightness_range=[0.7, 1.3],
        shear_range=0.2,
        fill_mode='nearest'
    )
    groups=df.groupby('labels')
    for label in df['labels'].unique():
        group=groups.get_group(label)
        sample_count=len(group)
        if sample_count< n:
            aug_img_count=0
            delta=n - sample_count
            target_dir=os.path.join(aug_dir, label)
            msg='{0:40s} for class {1:^30s} creating {2:^5s} augmented images'.format(' ', label, str(delta))
            print(msg, '\r', end='')
            aug_gen=gen.flow_from_dataframe(group, x_col='filepaths', y_col=None, target_size=img_size,
                                            class_mode=None, batch_size=1, shuffle=False,
                                            save_to_dir=target_dir, save_prefix='aug-', color_mode='rgb',
                                            save_format='jpg')
            while aug_img_count<delta:
                images=next(aug_gen)
                aug_img_count += len(images)
            total +=aug_img_count
    print('Total Augmented images created= ', total)
    
    aug_fpaths=[]
    aug_labels=[]
    classlist=os.listdir(aug_dir)
    for klass in classlist:
        classpath=os.path.join(aug_dir, klass)
        flist=os.listdir(classpath)
        for f in flist:
            fpath=os.path.join(classpath,f)
            aug_fpaths.append(fpath)
            aug_labels.append(klass)
    Fseries=pd.Series(aug_fpaths, name='filepaths')
    Lseries=pd.Series(aug_labels, name='labels')
    aug_df=pd.concat([Fseries, Lseries], axis=1)
    df=pd.concat([df,aug_df], axis=0).reset_index(drop=True)
    print('Length of augmented dataframe is now ', len(df))
    return df

n=600  # Increased for better balance
working_dir=save_dir  # use /kaggle/working
img_size=(224,224)
train_df=balance(train_df, n, working_dir, img_size)

# Cell 6: Relabeling
def relabel(label_in):
    if label_in in ['Healthy', 'Doubtful', 'Minimal']:
        label='Healthy'
    else:
        label=label_in
    return label

# Use vectorized apply instead of chained assignment
for name, df in [('train', train_df), ('test', test_df), ('valid', valid_df)]:
    df_copy = df.copy()
    df_copy['labels'] = df_copy['labels'].apply(relabel)
    if name == 'train':
        train_df = df_copy
    elif name == 'test':
        test_df = df_copy
    else:
        valid_df = df_copy
        
print(train_df['labels'].value_counts())

# Cell 7: Final trimming
max_samples=600
min_samples=600
column='labels'
train_df= trim(train_df, max_samples, min_samples, column)

# Cell 8: Enhanced Data generators with VGG16 preprocessing
batch_size=16  # Optimal batch size for VGG16

# Strong augmentation for training
trgen=ImageDataGenerator(
    horizontal_flip=True,
    rotation_range=30,
    width_shift_range=0.25,
    height_shift_range=0.25,
    zoom_range=0.25,
    brightness_range=[0.7, 1.3],
    shear_range=0.2,
    channel_shift_range=20,
    fill_mode='nearest',
    preprocessing_function=tf.keras.applications.vgg16.preprocess_input
)

# Validation and test use only preprocessing
t_and_v_gen=ImageDataGenerator(
    preprocessing_function=tf.keras.applications.vgg16.preprocess_input
)

msg='{0:70s} for train generator'.format(' ')
print(msg, '\r', end='')
train_gen=trgen.flow_from_dataframe(train_df, x_col='filepaths', y_col='labels', target_size=img_size,
                                   class_mode='categorical', color_mode='rgb', shuffle=True, batch_size=batch_size)
msg='{0:70s} for valid generator'.format(' ')
print(msg, '\r', end='')
valid_gen=t_and_v_gen.flow_from_dataframe(valid_df, x_col='filepaths', y_col='labels', target_size=img_size,
                                   class_mode='categorical', color_mode='rgb', shuffle=False, batch_size=batch_size)

length=len(test_df)
test_batch_size=sorted([int(length/n) for n in range(1,length+1) if length % n ==0 and length/n<=80],reverse=True)[0]
test_steps=int(length/test_batch_size)
msg='{0:70s} for test generator'.format(' ')
print(msg, '\r', end='')
test_gen=t_and_v_gen.flow_from_dataframe(test_df, x_col='filepaths', y_col='labels', target_size=img_size,
                                   class_mode='categorical', color_mode='rgb', shuffle=False, batch_size=test_batch_size)

classes=list(train_gen.class_indices.keys())
class_indices=list(train_gen.class_indices.values())
class_count=len(classes)
labels=test_gen.labels
print('test batch size: ', test_batch_size, '  test steps: ', test_steps, ' number of classes : ', class_count)

# Cell 9: Advanced VGG16 Model with Multiple Dropout Layers
def create_advanced_vgg16_model(img_size, class_count):
    """
    Creates an advanced VGG16 model with:
    - Dual pooling (GAP + GMP) for robust feature extraction
    - Multiple dropout layers at different rates
    - Batch normalization for stable training
    - L2 regularization to prevent overfitting
    - Dense layers with decreasing neurons
    """
    input_layer = Input(shape=(img_size[0], img_size[1], 3), name='input_layer')
    
    # Load VGG16 with ImageNet weights
    base_model = VGG16(weights='imagenet', include_top=False, input_tensor=input_layer)
    
    # Dual pooling strategy - captures both average and max features
    gap = GlobalAveragePooling2D(name='global_avg_pool')(base_model.output)
    gmp = GlobalMaxPooling2D(name='global_max_pool')(base_model.output)
    
    # Concatenate both pooling outputs
    concat_pool = concatenate([gap, gmp], name='concat_pooling')
    concat_pool = BatchNormalization(name='bn_concat')(concat_pool)
    concat_pool = Dropout(0.6, name='dropout_concat')(concat_pool)  # High dropout after pooling
    
    # First dense block - 1024 neurons
    x = Dense(1024, kernel_regularizer=regularizers.l2(0.002), name='dense_1')(concat_pool)
    x = BatchNormalization(name='bn_1')(x)
    x = Activation('relu', name='relu_1')(x)
    x = Dropout(0.55, name='dropout_1')(x)  # 55% dropout
    
    # Second dense block - 512 neurons
    x = Dense(512, kernel_regularizer=regularizers.l2(0.002), name='dense_2')(x)
    x = BatchNormalization(name='bn_2')(x)
    x = Activation('relu', name='relu_2')(x)
    x = Dropout(0.5, name='dropout_2')(x)  # 50% dropout
    
    # Third dense block - 256 neurons
    x = Dense(256, kernel_regularizer=regularizers.l2(0.001), name='dense_3')(x)
    x = BatchNormalization(name='bn_3')(x)
    x = Activation('relu', name='relu_3')(x)
    x = Dropout(0.45, name='dropout_3')(x)  # 45% dropout
    
    # Fourth dense block - 128 neurons
    x = Dense(128, kernel_regularizer=regularizers.l2(0.001), name='dense_4')(x)
    x = BatchNormalization(name='bn_4')(x)
    x = Activation('relu', name='relu_4')(x)
    x = Dropout(0.4, name='dropout_4')(x)  # 40% dropout
    
    # Output layer
    predictions = Dense(class_count, activation='softmax', name='predictions')(x)
    
    model = Model(inputs=input_layer, outputs=predictions)
    
    return model, base_model

# Create the VGG16 model
print("\n" + "="*80)
print("Creating Advanced VGG16 Model with Multiple Dropout Layers")
print("="*80)
model, base_model = create_advanced_vgg16_model(img_size, class_count)

# Print model architecture
model.summary()
print(f"\nTotal Dropout Layers: 5")
print(f"Dropout Rates: 60%, 55%, 50%, 45%, 40%")
print(f"Total Trainable Parameters: {model.count_params():,}")

# Cell 10: Custom Learning Rate Schedule
def lr_schedule(epoch, lr):
    """
    Custom learning rate schedule for optimal convergence
    """
    if epoch < 10:
        return 0.001
    elif epoch < 25:
        return 0.0005
    elif epoch < 40:
        return 0.0001
    elif epoch < 60:
        return 0.00005
    else:
        return 0.00001

# Cell 11: Phase 1 - Train Top Layers Only
print("\n" + "="*80)
print("PHASE 1: Training Top Layers (Frozen VGG16 Base)")
print("="*80)

# Freeze all VGG16 layers
for layer in base_model.layers:
    layer.trainable = False

# Compile model
model.compile(
    optimizer=Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999),
    loss='categorical_crossentropy',
    metrics=['accuracy', tf.keras.metrics.Precision(name='precision'), 
             tf.keras.metrics.Recall(name='recall'),
             tf.keras.metrics.AUC(name='auc')]
)

# Callbacks for Phase 1
callbacks_phase1 = [
    EarlyStopping(
        monitor='val_loss', 
        patience=12, 
        restore_best_weights=True, 
        verbose=1,
        mode='min'
    ),
    ReduceLROnPlateau(
        monitor='val_loss', 
        factor=0.5, 
        patience=6, 
        min_lr=1e-8, 
        verbose=1,
        mode='min'
    ),
    ModelCheckpoint(
        filepath=os.path.join(save_dir, 'vgg16_phase1_best.weights.h5'),
        monitor='val_accuracy',
        save_best_only=True,
        save_weights_only=True,
        verbose=1,
        mode='max'
    )
]

# Train Phase 1
print("\nTraining Phase 1...")
history_phase1 = model.fit(
    train_gen,
    steps_per_epoch=len(train_gen),
    epochs=30,
    validation_data=valid_gen,
    validation_steps=len(valid_gen),
    callbacks=callbacks_phase1,
    verbose=1
)

# Cell 12: Phase 2 - Fine-tune Last Conv Block
print("\n" + "="*80)
print("PHASE 2: Fine-tuning Last Conv Block (Block5)")
print("="*80)

# Unfreeze block5 of VGG16 (last convolutional block)
for layer in base_model.layers:
    if 'block5' in layer.name:
        layer.trainable = True
    else:
        layer.trainable = False

# Count trainable parameters
trainable_count = sum([K.count_params(w) for w in model.trainable_weights])
print(f"Trainable parameters: {trainable_count:,}")

# Recompile with lower learning rate
model.compile(
    optimizer=Adam(learning_rate=0.0002, beta_1=0.9, beta_2=0.999),
    loss='categorical_crossentropy',
    metrics=['accuracy', tf.keras.metrics.Precision(name='precision'), 
             tf.keras.metrics.Recall(name='recall'),
             tf.keras.metrics.AUC(name='auc')]
)

# Callbacks for Phase 2
callbacks_phase2 = [
    EarlyStopping(
        monitor='val_loss', 
        patience=15, 
        restore_best_weights=True, 
        verbose=1,
        mode='min'
    ),
    ReduceLROnPlateau(
        monitor='val_loss', 
        factor=0.4, 
        patience=8, 
        min_lr=1e-9, 
        verbose=1,
        mode='min'
    ),
    ModelCheckpoint(
        filepath=os.path.join(save_dir, 'vgg16_phase2_best.weights.h5'),
        monitor='val_accuracy',
        save_best_only=True,
        save_weights_only=True,
        verbose=1,
        mode='max'
    ),
    LearningRateScheduler(lr_schedule, verbose=1)
]

# Train Phase 2
print("\nTraining Phase 2...")
history_phase2 = model.fit(
    train_gen,
    steps_per_epoch=len(train_gen),
    epochs=35,
    validation_data=valid_gen,
    validation_steps=len(valid_gen),
    callbacks=callbacks_phase2,
    verbose=1
)

# Cell 13: Phase 3 - Fine-tune Block4 and Block5
print("\n" + "="*80)
print("PHASE 3: Fine-tuning Block4 and Block5")
print("="*80)

# Unfreeze block4 and block5
for layer in base_model.layers:
    if 'block4' in layer.name or 'block5' in layer.name:
        layer.trainable = True
    else:
        layer.trainable = False

trainable_count = sum([K.count_params(w) for w in model.trainable_weights])
print(f"Trainable parameters: {trainable_count:,}")

# Recompile with even lower learning rate
model.compile(
    optimizer=Adam(learning_rate=0.00005, beta_1=0.9, beta_2=0.999),
    loss='categorical_crossentropy',
    metrics=['accuracy', tf.keras.metrics.Precision(name='precision'), 
             tf.keras.metrics.Recall(name='recall'),
             tf.keras.metrics.AUC(name='auc')]
)

callbacks_phase3 = [
    EarlyStopping(
        monitor='val_loss', 
        patience=20, 
        restore_best_weights=True, 
        verbose=1,
        mode='min'
    ),
    ReduceLROnPlateau(
        monitor='val_loss', 
        factor=0.3, 
        patience=10, 
        min_lr=1e-10, 
        verbose=1,
        mode='min'
    ),
    ModelCheckpoint(
        filepath=os.path.join(save_dir, 'vgg16_phase3_best.weights.h5'),
        monitor='val_accuracy',
        save_best_only=True,
        save_weights_only=True,
        verbose=1,
        mode='max'
    )
]

# Train Phase 3
print("\nTraining Phase 3...")
history_phase3 = model.fit(
    train_gen,
    steps_per_epoch=len(train_gen),
    epochs=40,
    validation_data=valid_gen,
    validation_steps=len(valid_gen),
    callbacks=callbacks_phase3,
    verbose=1
)

# Cell 14: Phase 4 - Full Model Fine-tuning (Optional for maximum accuracy)
print("\n" + "="*80)
print("PHASE 4: Full Model Fine-tuning with Ultra-Low Learning Rate")
print("="*80)

# Unfreeze all layers
for layer in model.layers:
    layer.trainable = True

trainable_count = sum([K.count_params(w) for w in model.trainable_weights])
print(f"Trainable parameters: {trainable_count:,}")

# Recompile with ultra-low learning rate
model.compile(
    optimizer=Adam(learning_rate=0.00001, beta_1=0.9, beta_2=0.999),
    loss='categorical_crossentropy',
    metrics=['accuracy', tf.keras.metrics.Precision(name='precision'), 
             tf.keras.metrics.Recall(name='recall'),
             tf.keras.metrics.AUC(name='auc')]
)

callbacks_phase4 = [
    EarlyStopping(
        monitor='val_loss', 
        patience=25, 
        restore_best_weights=True, 
        verbose=1,
        mode='min'
    ),
    ReduceLROnPlateau(
        monitor='val_loss', 
        factor=0.2, 
        patience=12, 
        min_lr=1e-11, 
        verbose=1,
        mode='min'
    ),
    ModelCheckpoint(
        filepath=os.path.join(save_dir, 'vgg16_phase4_best.weights.h5'),
        monitor='val_accuracy',
        save_best_only=True,
        save_weights_only=True,
        verbose=1,
        mode='max'
    )
]

# Train Phase 4
print("\nTraining Phase 4...")
history_phase4 = model.fit(
    train_gen,
    steps_per_epoch=len(train_gen),
    epochs=35,
    validation_data=valid_gen,
    validation_steps=len(valid_gen),
    callbacks=callbacks_phase4,
    verbose=1
)

# Cell 15: Final Evaluation
print("\n" + "="*80)
print("FINAL MODEL EVALUATION")
print("="*80)

# Evaluate on test set
print("\nEvaluating on test set...")
result = model.evaluate(test_gen, steps=test_steps, verbose=1)

print("\n" + "="*80)
print("FINAL RESULTS:")
print("="*80)
print(f"Test Loss:      {result[0]:.5f}")
print(f"Test Accuracy:  {result[1]*100:.2f}%")
print(f"Test Precision: {result[2]*100:.2f}%")
print(f"Test Recall:    {result[3]*100:.2f}%")
print(f"Test AUC:       {result[4]:.5f}")
print("="*80)

# Get predictions
print("\nGenerating predictions...")
test_gen.reset()
y_pred_proba = model.predict(test_gen, steps=test_steps, verbose=1)
y_pred = np.argmax(y_pred_proba, axis=1)
y_true = test_gen.classes

# Cell 16: Detailed Classification Report
print("\n" + "="*80)
print("CLASSIFICATION REPORT:")
print("="*80)
print(classification_report(y_true, y_pred, target_names=classes, digits=4))

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(12, 10))
sns.heatmap(cm, annot=True, fmt='d', cmap='YlOrRd', xticklabels=classes, yticklabels=classes, 
            cbar_kws={'label': 'Count'}, annot_kws={'size': 14})
plt.title('Confusion Matrix - VGG16 Model\n', fontsize=16, fontweight='bold')
plt.ylabel('True Label', fontsize=12)
plt.xlabel('Predicted Label', fontsize=12)
plt.tight_layout()
plt.savefig(os.path.join(save_dir, 'vgg16_confusion_matrix.png'), dpi=300, bbox_inches='tight')
plt.show()

# Cell 17: Comprehensive Training History Visualization
fig = plt.figure(figsize=(20, 12))

# Combine all histories
all_accuracy = (history_phase1.history['accuracy'] + history_phase2.history['accuracy'] + 
                history_phase3.history['accuracy'] + history_phase4.history['accuracy'])
all_val_accuracy = (history_phase1.history['val_accuracy'] + history_phase2.history['val_accuracy'] + 
                    history_phase3.history['val_accuracy'] + history_phase4.history['val_accuracy'])
all_loss = (history_phase1.history['loss'] + history_phase2.history['loss'] + 
            history_phase3.history['loss'] + history_phase4.history['loss'])
all_val_loss = (history_phase1.history['val_loss'] + history_phase2.history['val_loss'] + 
                history_phase3.history['val_loss'] + history_phase4.history['val_loss'])

# Plot 1: Accuracy
ax1 = plt.subplot(2, 3, 1)
epochs_range = range(1, len(all_accuracy) + 1)
ax1.plot(epochs_range, all_accuracy, 'b-', label='Training Accuracy', linewidth=2)
ax1.plot(epochs_range, all_val_accuracy, 'r-', label='Validation Accuracy', linewidth=2)
ax1.axhline(y=0.95, color='g', linestyle='--', label='Target: 95%', linewidth=1.5)
ax1.set_title('Model Accuracy Over Time', fontsize=14, fontweight='bold')
ax1.set_xlabel('Epoch', fontsize=11)
ax1.set_ylabel('Accuracy', fontsize=11)
ax1.legend(loc='lower right')
ax1.grid(True, alpha=0.3)
ax1.set_ylim([0, 1.05])

# Plot 2: Loss
ax2 = plt.subplot(2, 3, 2)
ax2.plot(epochs_range, all_loss, 'b-', label='Training Loss', linewidth=2)
ax2.plot(epochs_range, all_val_loss, 'r-', label='Validation Loss', linewidth=2)
ax2.axhline(y=0.5, color='g', linestyle='--', label='Target: 0.5', linewidth=1.5)
ax2.set_title('Model Loss Over Time', fontsize=14, fontweight='bold')
ax2.set_xlabel('Epoch', fontsize=11)
ax2.set_ylabel('Loss', fontsize=11)
ax2.legend(loc='upper right')
ax2.grid(True, alpha=0.3)

# Plot 3: ROC Curves
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
from itertools import cycle

ax3 = plt.subplot(2, 3, 3)
y_true_bin = label_binarize(y_true, classes=list(range(class_count)))
fpr = dict()
tpr = dict()
roc_auc = dict()

colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A', '#98D8C8']
for i in range(class_count):
    fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_pred_proba[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
    ax3.plot(fpr[i], tpr[i], color=colors[i % len(colors)], lw=2.5,
             label=f'{classes[i]} (AUC = {roc_auc[i]:.3f})')

ax3.plot([0, 1], [0, 1], 'k--', lw=2, label='Random Classifier')
ax3.set_xlim([0.0, 1.0])
ax3.set_ylim([0.0, 1.05])
ax3.set_xlabel('False Positive Rate', fontsize=11)
ax3.set_ylabel('True Positive Rate', fontsize=11)
ax3.set_title('ROC Curves (Multi-class)', fontsize=14, fontweight='bold')
ax3.legend(loc="lower right", fontsize=9)
ax3.grid(True, alpha=0.3)

# Plot 4: Per-Class Accuracy
ax4 = plt.subplot(2, 3, 4)
class_accuracy = []
for i in range(class_count):
    class_mask = y_true == i
    if np.sum(class_mask) > 0:
        class_acc = np.mean(y_pred[class_mask] == y_true[class_mask]) * 100
    else:
        class_acc = 0
    class_accuracy.append(class_acc)

bars = ax4.bar(classes, class_accuracy, color=colors[:class_count], edgecolor='black', linewidth=1.5)
ax4.axhline(y=95, color='green', linestyle='--', label='Target: 95%', linewidth=2)
ax4.set_title('Per-Class Accuracy', fontsize=14, fontweight='bold')
ax4.set_ylabel('Accuracy (%)', fontsize=11)
ax4.set_ylim([0, 105])
for i, (bar, v) in enumerate(zip(bars, class_accuracy)):
    height = bar.get_height()
    ax4.text(bar.get_x() + bar.get_width()/2., height + 1,
             f'{v:.1f}%', ha='center', va='bottom', fontweight='bold', fontsize=10)
ax4.legend()
ax4.grid(True, axis='y', alpha=0.3)

# Plot 5: Precision-Recall per Class
from sklearn.metrics import precision_recall_curve, average_precision_score

ax5 = plt.subplot(2, 3, 5)
for i in range(class_count):
    precision, recall, _ = precision_recall_curve(y_true_bin[:, i], y_pred_proba[:, i])
    avg_precision = average_precision_score(y_true_bin[:, i], y_pred_proba[:, i])
    ax5.plot(recall, precision, color=colors[i % len(colors)], lw=2.5,
             label=f'{classes[i]} (AP = {avg_precision:.3f})')

ax5.set_xlabel('Recall', fontsize=11)
ax5.set_ylabel('Precision', fontsize=11)
ax5.set_title('Precision-Recall Curves', fontsize=14, fontweight='bold')
ax5.legend(loc="lower left", fontsize=9)
ax5.grid(True, alpha=0.3)
ax5.set_xlim([0.0, 1.0])
ax5.set_ylim([0.0, 1.05])

# Plot 6: Training Metrics Summary
ax6 = plt.subplot(2, 3, 6)
metrics_names = ['Accuracy', 'Precision', 'Recall', 'AUC']
metrics_values = [result[1]*100, result[2]*100, result[3]*100, result[4]*100]
bars2 = ax6.barh(metrics_names, metrics_values, color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A'], 
                 edgecolor='black', linewidth=1.5)
ax6.axvline(x=95, color='green', linestyle='--', label='Target: 95%', linewidth=2)
ax6.set_xlabel('Score (%)', fontsize=11)
ax6.set_title('Final Test Metrics Summary', fontsize=14, fontweight='bold')
ax6.set_xlim([0, 105])
for i, (bar, v) in enumerate(zip(bars2, metrics_values)):
    width = bar.get_width()
    ax6.text(width + 1, bar.get_y() + bar.get_height()/2.,
             f'{v:.2f}%', ha='left', va='center', fontweight='bold', fontsize=11)
ax6.legend()
ax6.grid(True, axis='x', alpha=0.3)

plt.tight_layout()
plt.savefig(os.path.join(save_dir, 'vgg16_training_results.png'), dpi=300, bbox_inches='tight')
plt.show()

# Cell 18: Detailed Performance Analysis
print("\n" + "="*80)
print("DETAILED PERFORMANCE ANALYSIS")
print("="*80)

# Calculate additional metrics
from sklearn.metrics import cohen_kappa_score, matthews_corrcoef

kappa = cohen_kappa_score(y_true, y_pred)
mcc = matthews_corrcoef(y_true, y_pred)

print(f"\nCohen's Kappa Score: {kappa:.4f}")
print(f"Matthews Correlation Coefficient: {mcc:.4f}")

# Per-class analysis
print("\n" + "-"*80)
print("PER-CLASS DETAILED ANALYSIS:")
print("-"*80)
print(f"{'Class':<15} {'Accuracy':<12} {'Precision':<12} {'Recall':<12} {'F1-Score':<12} {'Support':<10}")
print("-"*80)

from sklearn.metrics import precision_score, recall_score, f1_score

for i, class_name in enumerate(classes):
    class_mask = y_true == i
    class_support = np.sum(class_mask)
    
    if class_support > 0:
        class_acc = np.mean(y_pred[class_mask] == y_true[class_mask]) * 100
        
        # Create binary arrays for this class
        y_true_binary = (y_true == i).astype(int)
        y_pred_binary = (y_pred == i).astype(int)
        
        class_precision = precision_score(y_true_binary, y_pred_binary, zero_division=0) * 100
        class_recall = recall_score(y_true_binary, y_pred_binary, zero_division=0) * 100
        class_f1 = f1_score(y_true_binary, y_pred_binary, zero_division=0) * 100
        
        print(f"{class_name:<15} {class_acc:>10.2f}% {class_precision:>10.2f}% {class_recall:>10.2f}% {class_f1:>10.2f}% {class_support:>10}")

print("-"*80)

# Cell 19: Learning Rate History Visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Reconstruct learning rate history
phase1_epochs = len(history_phase1.history['loss'])
phase2_epochs = len(history_phase2.history['loss'])
phase3_epochs = len(history_phase3.history['loss'])
phase4_epochs = len(history_phase4.history['loss'])

total_epochs = phase1_epochs + phase2_epochs + phase3_epochs + phase4_epochs
epoch_numbers = list(range(1, total_epochs + 1))

# Create learning rate timeline
lr_history = []
# Phase 1
lr_history.extend([0.001] * phase1_epochs)
# Phase 2
lr_history.extend([0.0002] * phase2_epochs)
# Phase 3
lr_history.extend([0.00005] * phase3_epochs)
# Phase 4
lr_history.extend([0.00001] * phase4_epochs)

ax1.plot(epoch_numbers, lr_history, 'b-', linewidth=2.5, marker='o', markersize=3)
ax1.set_xlabel('Epoch', fontsize=12)
ax1.set_ylabel('Learning Rate', fontsize=12)
ax1.set_title('Learning Rate Schedule', fontsize=14, fontweight='bold')
ax1.set_yscale('log')
ax1.grid(True, alpha=0.3, which="both")
ax1.axvline(x=phase1_epochs, color='r', linestyle='--', alpha=0.7, label='Phase 1 End')
ax1.axvline(x=phase1_epochs+phase2_epochs, color='g', linestyle='--', alpha=0.7, label='Phase 2 End')
ax1.axvline(x=phase1_epochs+phase2_epochs+phase3_epochs, color='orange', linestyle='--', alpha=0.7, label='Phase 3 End')
ax1.legend()

# Accuracy improvement across phases
phase_boundaries = [0, phase1_epochs, phase1_epochs+phase2_epochs, 
                   phase1_epochs+phase2_epochs+phase3_epochs, total_epochs]
phase_names = ['Phase 1', 'Phase 2', 'Phase 3', 'Phase 4']
phase_colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A']

for i in range(4):
    start_idx = phase_boundaries[i]
    end_idx = phase_boundaries[i+1]
    phase_epochs = epoch_numbers[start_idx:end_idx]
    phase_acc = all_val_accuracy[start_idx:end_idx]
    ax2.plot(phase_epochs, phase_acc, color=phase_colors[i], linewidth=3, 
             label=phase_names[i], marker='o', markersize=4)

ax2.axhline(y=0.95, color='green', linestyle='--', linewidth=2, label='Target: 95%')
ax2.set_xlabel('Epoch', fontsize=12)
ax2.set_ylabel('Validation Accuracy', fontsize=12)
ax2.set_title('Validation Accuracy by Training Phase', fontsize=14, fontweight='bold')
ax2.legend(loc='lower right')
ax2.grid(True, alpha=0.3)
ax2.set_ylim([0, 1.05])

plt.tight_layout()
plt.savefig(os.path.join(save_dir, 'vgg16_learning_analysis.png'), dpi=300, bbox_inches='tight')
plt.show()

# Cell 20: Model Architecture Summary
print("\n" + "="*80)
print("MODEL ARCHITECTURE SUMMARY")
print("="*80)
print("\nVGG16 Base Model:")
print(f"  - Total Layers: {len(base_model.layers)}")
print(f"  - Convolutional Blocks: 5 (Block1 to Block5)")
print(f"  - Each block has 2-3 Conv2D layers with MaxPooling")

print("\nCustom Classification Head:")
print(f"  - Dual Pooling: GlobalAveragePooling2D + GlobalMaxPooling2D")
print(f"  - Dense Layer 1: 1024 neurons + BatchNorm + Dropout(0.55)")
print(f"  - Dense Layer 2: 512 neurons + BatchNorm + Dropout(0.50)")
print(f"  - Dense Layer 3: 256 neurons + BatchNorm + Dropout(0.45)")
print(f"  - Dense Layer 4: 128 neurons + BatchNorm + Dropout(0.40)")
print(f"  - Output Layer: {class_count} neurons (Softmax)")

print(f"\nTotal Dropout Layers: 5")
print(f"Dropout Rates: 60%, 55%, 50%, 45%, 40%")
print(f"L2 Regularization: 0.001-0.002")

print("\nTraining Strategy:")
print(f"  - Phase 1: Top layers only (30 epochs)")
print(f"  - Phase 2: + Block5 fine-tuning (35 epochs)")
print(f"  - Phase 3: + Block4 fine-tuning (40 epochs)")
print(f"  - Phase 4: Full model fine-tuning (35 epochs)")
print(f"  - Total Epochs: {total_epochs}")

total_params = model.count_params()
trainable_params = sum([K.count_params(w) for w in model.trainable_weights])
non_trainable_params = total_params - trainable_params

print(f"\nModel Parameters:")
print(f"  - Total Parameters: {total_params:,}")
print(f"  - Trainable Parameters: {trainable_params:,}")
print(f"  - Non-trainable Parameters: {non_trainable_params:,}")

# Cell 21: Save Final Model (robust)
print("\n" + "="*80)
print("SAVING MODEL")
print("="*80)

# Try to load best phase 4 weights before final save
best_phase4_weights = os.path.join(save_dir, 'vgg16_phase4_best.weights.h5')
if os.path.exists(best_phase4_weights):
    try:
        model.load_weights(best_phase4_weights)
        print(f"Loaded best Phase 4 weights from: {best_phase4_weights}")
    except Exception as e:
        print("⚠ Failed to load best Phase 4 weights:", repr(e))
else:
    print("No Phase 4 best weights file found; saving current model weights.")

# Show disk usage
du = shutil.disk_usage(save_dir)
print(f"Disk usage for {save_dir} -> total: {du.total//(1024**3)}GB, used: {du.used//(1024**3)}GB, free: {du.free//(1024**3)}GB")

# Save as SavedModel directory
savedmodel_dir = os.path.join(save_dir, 'vgg16_knee_osteoarthritis_final_tf')
h5_file = os.path.join(save_dir, 'vgg16_knee_osteoarthritis_final.h5')

try:
    model.save(savedmodel_dir, save_format='tf')
    print(f"✓ Full model saved in SavedModel format at: {savedmodel_dir}")
except Exception as e:
    print("⚠ Failed to save SavedModel format:", repr(e))

# Save HDF5 model (no optimizer to reduce size)
try:
    model.save(h5_file, include_optimizer=False)
    print(f"✓ Full model saved as HDF5 at: {h5_file}")
except Exception as e:
    print("⚠ Failed to save HDF5 model:", repr(e))

# Save weights explicitly
try:
    weights_file = os.path.join(save_dir, 'vgg16_weights_final.h5')
    model.save_weights(weights_file)
    print(f"✓ Model weights saved as: {weights_file}")
except Exception as e:
    print("⚠ Failed to save weights:", repr(e))

# Save model architecture as JSON
try:
    model_json = model.to_json()
    json_path = os.path.join(save_dir, 'vgg16_architecture.json')
    with open(json_path, 'w') as json_file:
        json_file.write(model_json)
    print(f"✓ Model architecture saved as: {json_path}")
except Exception as e:
    print("⚠ Failed to save model JSON:", repr(e))

# Save training history
try:
    import pickle
    history_dict = {
        'phase1': history_phase1.history,
        'phase2': history_phase2.history,
        'phase3': history_phase3.history,
        'phase4': history_phase4.history
    }
    history_path = os.path.join(save_dir, 'training_history.pkl')
    with open(history_path, 'wb') as file:
        pickle.dump(history_dict, file)
    print(f"✓ Training history saved as: {history_path}")
except Exception as e:
    print("⚠ Failed to save training history:", repr(e))

# Cell 22: Final Summary Report
print("\n" + "="*80)
print("FINAL TRAINING SUMMARY REPORT")
print("="*80)

print("\n📊 DATASET STATISTICS:")
print(f"  - Training Samples: {len(train_df):,}")
print(f"  - Validation Samples: {len(valid_df):,}")
print(f"  - Test Samples: {len(test_df):,}")
print(f"  - Number of Classes: {class_count}")
print(f"  - Classes: {', '.join(classes)}")

print("\n🎯 TARGET METRICS:")
print(f"  - Target Accuracy: 95.00%")
print(f"  - Target Loss: 0.50")

print("\n📈 ACHIEVED METRICS:")
print(f"  - Test Accuracy: {result[1]*100:.2f}%")
print(f"  - Test Loss: {result[0]:.5f}")
print(f"  - Test Precision: {result[2]*100:.2f}%")
print(f"  - Test Recall: {result[3]*100:.2f}%")
print(f"  - Test AUC: {result[4]:.4f}")

accuracy_diff = (result[1]*100) - 95.0
loss_diff = result[0] - 0.5

print("\n📊 TARGET COMPARISON:")
if result[1]*100 >= 95.0:
    print(f"  ✓ Accuracy: ACHIEVED (exceeded by {accuracy_diff:.2f}%)")
else:
    print(f"  ✗ Accuracy: CLOSE (short by {abs(accuracy_diff):.2f}%)")
    
if result[0] <= 0.5:
    print(f"  ✓ Loss: ACHIEVED (better by {abs(loss_diff):.4f})")
else:
    print(f"  ✗ Loss: CLOSE (higher by {loss_diff:.4f})")

print("\n🔧 MODEL CONFIGURATION:")
print(f"  - Architecture: VGG16 + Custom Head")
print(f"  - Input Size: {img_size}")
print(f"  - Batch Size: {batch_size}")
print(f"  - Total Epochs Trained: {total_epochs}")
print(f"  - Dropout Layers: 5 (rates: 60%, 55%, 50%, 45%, 40%)")
print(f"  - Regularization: L2 (0.001-0.002)")
print(f"  - Optimizer: Adam with adaptive learning rate")

print("\n💡 KEY FEATURES:")
print("  ✓ Dual pooling (GAP + GMP) for robust features")
print("  ✓ Progressive fine-tuning (4 phases)")
print("  ✓ Multiple dropout layers for regularization")
print("  ✓ Batch normalization for stable training")
print("  ✓ Enhanced data augmentation")
print("  ✓ Early stopping and LR scheduling")
print("  ✓ Model checkpointing for best weights")

print("\n📁 SAVED FILES:")
print("  - vgg16_knee_osteoarthritis_final_tf/ (SavedModel directory)")
print("  - vgg16_knee_osteoarthritis_final.h5")
print("  - vgg16_weights_final.h5")
print("  - vgg16_architecture.json")
print("  - training_history.pkl")
print("  - vgg16_training_results.png")
print("  - vgg16_learning_analysis.png")
print("  - vgg16_confusion_matrix.png")

print("\n" + "="*80)
print("✨ TRAINING COMPLETE! ✨")
print("="*80)

# Cell 23: Sample Predictions Visualization
print("\nGenerating sample predictions visualization...")

# Get a batch of test images
test_gen.reset()
test_images, test_labels_batch = next(test_gen)
predictions = model.predict(test_images)

# Plot sample predictions
fig, axes = plt.subplots(4, 5, figsize=(20, 16))
axes = axes.ravel()

for idx in range(min(20, len(test_images))):
    ax = axes[idx]
    
    # Display image
    img = test_images[idx]
    # Reverse VGG16 preprocessing for visualization
    img = img.copy()
    img[:, :, 0] += 103.939
    img[:, :, 1] += 116.779
    img[:, :, 2] += 123.68
    img = img[:, :, ::-1]  # BGR to RGB
    img = np.clip(img, 0, 255).astype('uint8')
    
    ax.imshow(img)
    
    # Get prediction
    true_class_idx = np.argmax(test_labels_batch[idx])
    pred_class_idx = np.argmax(predictions[idx])
    confidence = predictions[idx][pred_class_idx] * 100
    
    true_label = classes[true_class_idx]
    pred_label = classes[pred_class_idx]
    
    # Set title color based on correctness
    color = 'green' if true_class_idx == pred_class_idx else 'red'
    
    ax.set_title(f'True: {true_label}\nPred: {pred_label}\nConf: {confidence:.1f}%', 
                 color=color, fontsize=10, fontweight='bold')
    ax.axis('off')

plt.tight_layout()
plt.savefig(os.path.join(save_dir, 'sample_predictions.png'), dpi=300, bbox_inches='tight')
plt.show()

print("\n✓ Sample predictions saved as: sample_predictions.png")
print("\nAll visualizations and models have been saved successfully!")
print("="*80)

# ==============================================================
# Cell 24: REAL-TIME AUTOTEST EVALUATION
# ==============================================================

print("\n" + "="*80)
print("🚀 REAL-TIME AUTOTEST EVALUATION")
print("="*80)

autotest_path = '/kaggle/input/knee-osteoarthritis-dataset-with-severity/autotest'

# Check if autotest folder exists
if not os.path.exists(autotest_path):
    raise FileNotFoundError("❌ 'autotest' folder not found! Please upload it next to train/test/val.")

# Load autotest dataframe
autotest_filepaths = []
autotest_labels = []
autotest_classes = os.listdir(autotest_path)
for klass in autotest_classes:
    class_dir = os.path.join(autotest_path, klass)
    if not os.path.isdir(class_dir):
        continue
    file_list = os.listdir(class_dir)
    for f in file_list:
        autotest_filepaths.append(os.path.join(class_dir, f))
        autotest_labels.append(klass)

autotest_df = pd.DataFrame({
    'filepaths': autotest_filepaths,
    'labels': autotest_labels
})

print(f"Found {len(autotest_df)} images across {len(autotest_classes)} classes in autotest set.")

# Preprocessing same as test set
autotest_gen = t_and_v_gen.flow_from_dataframe(
    autotest_df,
    x_col='filepaths',
    y_col='labels',
    target_size=img_size,
    class_mode='categorical',
    color_mode='rgb',
    shuffle=False,
    batch_size=test_batch_size
)

# Load best trained model weights (Phase 4 best)
autotest_weights_path = os.path.join(save_dir, 'vgg16_phase4_best.weights.h5')

if os.path.exists(autotest_weights_path):
    try:
        model.load_weights(autotest_weights_path)
        print("✅ Loaded best weights for evaluation from:", autotest_weights_path)
    except Exception as e:
        print("⚠ Failed to load best weights:", repr(e))
        print("Using current model weights for autotest evaluation instead.")
else:
    print("⚠ Best weights file not found at", autotest_weights_path)
    print("Using current model weights for autotest evaluation.")

# Evaluate on autotest data
print("\nEvaluating on autotest set...")
auto_result = model.evaluate(autotest_gen, verbose=1)

print("\n" + "="*80)
print("AUTOTEST RESULTS")
print("="*80)
print(f"Autotest Loss:      {auto_result[0]:.5f}")
print(f"Autotest Accuracy:  {auto_result[1]*100:.2f}%")
print(f"Autotest Precision: {auto_result[2]*100:.2f}%")
print(f"Autotest Recall:    {auto_result[3]*100:.2f}%")
print(f"Autotest AUC:       {auto_result[4]:.5f}")
print("="*80)

# Generate predictions
autotest_gen.reset()
y_auto_pred_proba = model.predict(autotest_gen, verbose=1)
y_auto_pred = np.argmax(y_auto_pred_proba, axis=1)
y_auto_true = autotest_gen.classes

# Save predictions
predicted_labels = [classes[i] for i in y_auto_pred]
true_labels = [classes[i] for i in y_auto_true]

autotest_df['true_label'] = true_labels
autotest_df['pred_label'] = predicted_labels
autotest_df['confidence'] = np.max(y_auto_pred_proba, axis=1)

autotest_csv_path = os.path.join(save_dir, 'autotest_predictions.csv')
autotest_df.to_csv(autotest_csv_path, index=False)
print(f"\n✅ Saved detailed predictions to: {autotest_csv_path}")

# Confusion Matrix
cm_auto = confusion_matrix(y_auto_true, y_auto_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(cm_auto, annot=True, fmt='d', cmap='coolwarm',
            xticklabels=classes, yticklabels=classes, cbar=False)
plt.title('Confusion Matrix - Autotest Set')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.tight_layout()
plt.savefig(os.path.join(save_dir, 'autotest_confusion_matrix.png'), dpi=300, bbox_inches='tight')
plt.show()

# Visualization: Top correct and incorrect examples
print("\nDisplaying sample autotest predictions...")
fig, axes = plt.subplots(4, 5, figsize=(20, 16))
axes = axes.ravel()

# Randomly pick 20 images
sample_df = autotest_df.sample(n=min(20, len(autotest_df)), random_state=42)
for i, (idx, row) in enumerate(sample_df.iterrows()):
    ax = axes[i]
    img = cv2.imread(row['filepaths'])
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    ax.imshow(img)
    color = 'green' if row['true_label'] == row['pred_label'] else 'red'
    ax.set_title(f"True: {row['true_label']}\nPred: {row['pred_label']}\nConf: {row['confidence']*100:.1f}%", 
                 color=color, fontsize=10, fontweight='bold')
    ax.axis('off')

plt.tight_layout()
plt.savefig(os.path.join(save_dir, 'autotest_sample_predictions.png'), dpi=300, bbox_inches='tight')
plt.show()

print("\n✓ Autotest sample predictions saved as: autotest_sample_predictions.png")
print("✅ Real-time autotest evaluation complete!")
print("="*80)
```

```python
# ==============================================================
# CELL 21: FIXED MODEL SAVING (Replace your Cell 21 with this)
# ==============================================================

print("\n" + "="*80)
print("SAVING MODEL")
print("="*80)

import os
import json
import pickle

# Ensure working directory exists
save_dir = '/kaggle/working/'
os.makedirs(save_dir, exist_ok=True)

# Store phase epoch counts for later use
phase1_epochs = len(history_phase1.history['loss'])
phase2_epochs = len(history_phase2.history['loss'])
phase3_epochs = len(history_phase3.history['loss'])
phase4_epochs = len(history_phase4.history['loss'])
total_epochs = phase1_epochs + phase2_epochs + phase3_epochs + phase4_epochs

saved_files = []

# Method 1: Save in TensorFlow SavedModel format (RECOMMENDED)
try:
    model_path = os.path.join(save_dir, 'vgg16_knee_osteoarthritis_savedmodel')
    model.save(model_path, save_format='tf')
    print(f"✓ Model saved in TensorFlow SavedModel format: {model_path}")
    saved_files.append('vgg16_knee_osteoarthritis_savedmodel/')
except Exception as e:
    print(f"⚠ TensorFlow SavedModel format failed: {e}")

# Method 2: Save as .keras format (New Keras format)
try:
    keras_path = os.path.join(save_dir, 'vgg16_knee_osteoarthritis_final.keras')
    model.save(keras_path)
    print(f"✓ Model saved as .keras format: {keras_path}")
    saved_files.append('vgg16_knee_osteoarthritis_final.keras')
except Exception as e:
    print(f"⚠ .keras format failed: {e}")

# Method 3: Save weights only (MOST RELIABLE)
try:
    weights_path = os.path.join(save_dir, 'vgg16_weights_final.h5')
    model.save_weights(weights_path)
    print(f"✓ Model weights saved: {weights_path}")
    saved_files.append('vgg16_weights_final.h5')
except Exception as e:
    print(f"❌ CRITICAL: Weights saving failed: {e}")

# Method 4: Save model architecture as JSON
try:
    architecture_path = os.path.join(save_dir, 'vgg16_architecture.json')
    model_json = model.to_json()
    with open(architecture_path, 'w') as json_file:
        json_file.write(model_json)
    print(f"✓ Model architecture saved: {architecture_path}")
    saved_files.append('vgg16_architecture.json')
except Exception as e:
    print(f"⚠ Architecture saving failed: {e}")

# Save model configuration
try:
    config_path = os.path.join(save_dir, 'model_config.json')
    model_config = {
        'input_shape': img_size,
        'num_classes': class_count,
        'classes': classes,
        'class_indices': train_gen.class_indices,
        'batch_size': batch_size,
        'total_epochs': total_epochs,
        'phase1_epochs': phase1_epochs,
        'phase2_epochs': phase2_epochs,
        'phase3_epochs': phase3_epochs,
        'phase4_epochs': phase4_epochs,
        'test_results': {
            'loss': float(result[0]),
            'accuracy': float(result[1]),
            'precision': float(result[2]),
            'recall': float(result[3]),
            'auc': float(result[4])
        }
    }
    with open(config_path, 'w') as f:
        json.dump(model_config, f, indent=4)
    print(f"✓ Model configuration saved: {config_path}")
    saved_files.append('model_config.json')
except Exception as e:
    print(f"⚠ Configuration saving failed: {e}")

# Save training history
try:
    history_path = os.path.join(save_dir, 'training_history.pkl')
    history_dict = {
        'phase1': history_phase1.history,
        'phase2': history_phase2.history,
        'phase3': history_phase3.history,
        'phase4': history_phase4.history
    }
    with open(history_path, 'wb') as file:
        pickle.dump(history_dict, file)
    print(f"✓ Training history saved: {history_path}")
    saved_files.append('training_history.pkl')
except Exception as e:
    print(f"⚠ History saving failed: {e}")

# Save class indices mapping
try:
    class_indices_path = os.path.join(save_dir, 'class_indices.json')
    with open(class_indices_path, 'w') as f:
        json.dump(train_gen.class_indices, f, indent=4)
    print(f"✓ Class indices saved: {class_indices_path}")
    saved_files.append('class_indices.json')
except Exception as e:
    print(f"⚠ Class indices saving failed: {e}")

# Save predictions on test set
try:
    predictions_path = os.path.join(save_dir, 'test_predictions.csv')
    test_predictions_df = pd.DataFrame({
        'true_label': [classes[i] for i in y_true],
        'predicted_label': [classes[i] for i in y_pred],
        'confidence': np.max(y_pred_proba, axis=1)
    })
    test_predictions_df.to_csv(predictions_path, index=False)
    print(f"✓ Test predictions saved: {predictions_path}")
    saved_files.append('test_predictions.csv')
except Exception as e:
    print(f"⚠ Predictions saving failed: {e}")

# List all saved files
print("\n" + "="*80)
print("📁 SAVED FILES IN /kaggle/working/:")
print("="*80)
all_files = os.listdir(save_dir)
for file in sorted(all_files):
    file_path = os.path.join(save_dir, file)
    if os.path.isfile(file_path):
        size = os.path.getsize(file_path) / (1024 * 1024)
        print(f"  ✓ {file:<50s} ({size:>8.2f} MB)")
    elif os.path.isdir(file_path):
        # Calculate directory size
        total_size = 0
        for dirpath, dirnames, filenames in os.walk(file_path):
            for f in filenames:
                fp = os.path.join(dirpath, f)
                total_size += os.path.getsize(fp)
        size = total_size / (1024 * 1024)
        print(f"  ✓ {file+'/':<50s} ({size:>8.2f} MB)")

print("="*80)

# Create a comprehensive zip file for easy download
try:
    print("\nCreating comprehensive zip file...")
    import zipfile
    
    zip_path = os.path.join(save_dir, 'vgg16_complete_model_package.zip')
    with zipfile.ZipFile(zip_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for file in all_files:
            file_path = os.path.join(save_dir, file)
            if file.endswith('.zip'):
                continue  # Don't include the zip itself
            if os.path.isfile(file_path):
                zipf.write(file_path, arcname=file)
            elif os.path.isdir(file_path):
                for root, dirs, files in os.walk(file_path):
                    for f in files:
                        full_path = os.path.join(root, f)
                        arc_name = os.path.join(file, os.path.relpath(full_path, file_path))
                        zipf.write(full_path, arcname=arc_name)
    
    zip_size = os.path.getsize(zip_path) / (1024 * 1024)
    print(f"✓ Complete package created: vgg16_complete_model_package.zip ({zip_size:.2f} MB)")
    print("\n  📦 This zip contains ALL model files and can be downloaded from Kaggle's Output tab")
except Exception as e:
    print(f"⚠ Zip creation failed: {e}")

print("\n" + "="*80)
print("✨ MODEL SAVING COMPLETE! ✨")
print("="*80)

# ==============================================================
# FUNCTION TO RELOAD THE MODEL
# ==============================================================

def load_saved_model(save_dir='/kaggle/working/'):
    """
    Function to reload the saved model
    Returns: loaded model or None if all methods fail
    """
    import tensorflow as tf
    from tensorflow import keras
    import json
    
    print("\n" + "="*80)
    print("LOADING SAVED MODEL")
    print("="*80)
    
    # Method 1: Try loading .keras format
    try:
        keras_path = os.path.join(save_dir, 'vgg16_knee_osteoarthritis_final.keras')
        if os.path.exists(keras_path):
            model = keras.models.load_model(keras_path)
            print("✓ Model loaded from .keras format")
            return model
    except Exception as e:
        print(f"⚠ .keras format loading failed: {e}")
    
    # Method 2: Try loading SavedModel format
    try:
        model_path = os.path.join(save_dir, 'vgg16_knee_osteoarthritis_savedmodel')
        if os.path.exists(model_path):
            model = keras.models.load_model(model_path)
            print("✓ Model loaded from SavedModel format")
            return model
    except Exception as e:
        print(f"⚠ SavedModel format loading failed: {e}")
    
    # Method 3: Rebuild from architecture + weights
    try:
        architecture_path = os.path.join(save_dir, 'vgg16_architecture.json')
        weights_path = os.path.join(save_dir, 'vgg16_weights_final.h5')
        
        if os.path.exists(architecture_path) and os.path.exists(weights_path):
            print("Rebuilding model from architecture and weights...")
            with open(architecture_path, 'r') as json_file:
                model_json = json_file.read()
            model = keras.models.model_from_json(model_json)
            model.load_weights(weights_path)
            print("✓ Model loaded from architecture + weights")
            return model
    except Exception as e:
        print(f"⚠ Architecture + weights loading failed: {e}")
    
    # Method 4: Recreate model using function and load weights
    try:
        weights_path = os.path.join(save_dir, 'vgg16_weights_final.h5')
        config_path = os.path.join(save_dir, 'model_config.json')
        
        if os.path.exists(weights_path) and os.path.exists(config_path):
            print("Recreating model architecture and loading weights...")
            with open(config_path, 'r') as f:
                config = json.load(f)
            
            # Recreate model using the same function
            recreated_model, _ = create_advanced_vgg16_model(
                tuple(config['input_shape']), 
                config['num_classes']
            )
            recreated_model.load_weights(weights_path)
            print("✓ Model recreated and weights loaded")
            return recreated_model
    except Exception as e:
        print(f"⚠ Model recreation + weights loading failed: {e}")
    
    print("❌ All loading methods failed")
    return None

# Test loading the model
print("\n" + "="*80)
print("TESTING MODEL RELOAD")
print("="*80)

loaded_model = load_saved_model()

if loaded_model is not None:
    print("\n✅ MODEL SUCCESSFULLY RELOADED!")
    print(f"   Model type: {type(loaded_model)}")
    print(f"   Input shape: {loaded_model.input_shape}")
    print(f"   Output shape: {loaded_model.output_shape}")
    print(f"   Total parameters: {loaded_model.count_params():,}")
    
    # Quick verification prediction
    try:
        print("\nVerifying model with a test prediction...")
        test_gen.reset()
        test_sample = next(test_gen)
        prediction = loaded_model.predict(test_sample[0][:1], verbose=0)
        print(f"✓ Prediction successful! Output shape: {prediction.shape}")
    except Exception as e:
        print(f"⚠ Prediction test failed: {e}")
else:
    print("\n❌ MODEL RELOAD FAILED")
    print("   Please check the error messages above")

print("="*80)

# ==============================================================
# DOWNLOAD INSTRUCTIONS
# ==============================================================

print("\n" + "="*80)
print("📥 HOW TO DOWNLOAD YOUR MODEL FROM KAGGLE")
print("="*80)

print("""
METHOD 1 - Download via Kaggle Interface (EASIEST):
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
1. Look at the RIGHT SIDEBAR in your Kaggle notebook
2. Click on the 'Output' or 'Data' tab
3. Find 'vgg16_complete_model_package.zip'
4. Click the DOWNLOAD button (⬇) next to it
5. Extract the zip file on your computer

This zip contains ALL model files in one package!

METHOD 2 - Download Individual Files:
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
1. In the Output tab, you'll see all individual files
2. Download the ones you need:
   - vgg16_weights_final.h5 (weights only)
   - model_config.json (configuration)
   - vgg16_architecture.json (architecture)
   - training_history.pkl (training history)

METHOD 3 - Use Kaggle API (from your local machine):
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
Install Kaggle CLI and run:
   kaggle kernels output [your-username]/[kernel-name] -p ./downloads

METHOD 4 - Mount Google Drive (if using Colab after Kaggle):
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
""")

print("""
from google.colab import drive
import shutil
drive.mount('/content/drive')
shutil.copy('/kaggle/working/vgg16_complete_model_package.zip', 
            '/content/drive/MyDrive/')
""")

print("="*80)

print("""
✅ WHAT YOU HAVE NOW:
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
• Complete trained model in multiple formats
• Model weights (can be loaded separately)
• Model architecture (JSON format)
• Training history (all 4 phases)
• Model configuration (hyperparameters, results)
• Class indices mapping
• Test predictions CSV
• Complete package in a single ZIP file

💡 TO USE THE MODEL LATER:
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
Use the load_saved_model() function provided above, or:

# Quick load:
model = keras.models.load_model('vgg16_knee_osteoarthritis_final.keras')

# Or load weights only:
model.load_weights('vgg16_weights_final.h5')
""")

print("="*80)
print("🎉 SAVING COMPLETE! Your model is ready to download! 🎉")
print("="*80)
```
